```python
import json

def load():
	"""
	Loads json data from files and returns a dictionary for each data type
	where the key is function_id and the value is the source code or comment
	"""
	src_path = 'functions.json'
	com_path = 'comments.json'

	with open(src_path, 'r') as fp:
		src = json.load(fp, parse_int=True)

	with open(com_path, 'r') as fp:
		com = json.load(fp)

	return (src, com)
```

```python
data, comm = load()
```

```python
FILE_JST='funcom_java_ast'
```

```python
#file for ASTs
FILE_JST='java_ast_new'

#get corresponding comments
FILE_COMM='java_comments_new'

#eliminate mysterious error cases
error_keys=[]

with open (FILE_JST, 'w+') as f1:
    with open (FILE_COMM, 'w+') as f2:
        for key in comm:

            #class inserted for syntactic correctness
            temp_java= 'class Test{'+ data[key] + '}'
            #print(key, temp_java)
            try:
                tree = javalang.parse.parse(temp_java)

            except:
                print(key)
                error_keys.append(key)
                
            else:
                tree_line=str(tree.children)
                tree_line=tree_line.replace('\n', ' ')
                tree_line=tree_line.replace('\t', ' ')

                line=str(comm[key])
                # replace line-breaks & indentations by special symbols to get the comment in 1 line
                line=line.replace('\n', 'DCNL ')
                line=line.replace('\t', 'DCSP ')
                try:
                    f1.write(tree_line+'\n')
                    f2.write(line+'\n')
                except:
                    #print(key)
                    error_keys.append(key)

                #print(key, end=" ")
```
